```python
'''
# Sales Data Analysis using Python

## Project Overview

This project analyzes sales transaction data using Python and Pandas. The objective is to perform data cleaning, exploratory data analysis (EDA), revenue analysis, and identify meaningful business insights from the sales dataset.

## Steps Performed

### 1. Data Loading

* Imported the sales dataset using Pandas.
* Examined dataset structure and records.

### 2. Data Exploration

* Checked dataset dimensions.
* Viewed sample records.
* Analyzed data types and summary statistics.

### 3. Data Cleaning

* Checked for missing values.
* Removed duplicate records.
* Validated dataset quality.

### 4. Revenue Analysis

* Calculated total revenue by category.
* Compared category-wise sales performance.

### 5. Data Sorting

* Sorted products based on category and total sales value.
* Identified high-performing products.

### 6. Correlation Analysis

* Analyzed relationships between quantity, unit price, and total revenue.

### Tools Used

* Python
* Pandas
* NumPy
* Matplotlib

### Outcome

###The project provides insights into category-wise sales performance and helps understand factors influencing revenue generation.
'''
```

```python

#LOADING THE DATASET:
import pandas as pd
df=pd.read_csv("sales.csv",sep='\t')
print(df.describe())
print(df.info())
print(df.head(5))
```

```python
#HANDLING MISSING VALUES:
print(df.isnull().sum)
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#TOTAL REVENUE:

print(df.columns)
total_sales=df.groupby("category")["total_price"].sum()
print(total_sales)
```

```python
#SORTING
sorted_df=df.sort_values(by=['category', 'total_price'], ascending=[True, False])
print(sorted_df)
```

```python
#CORRELATION :

correlation=df.corr(numeric_only=True)
print(correlation)

import matplotlib.pyplot as plt

plt.scatter(df['quantity'], df['total_price'])
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.title("Quantity vs Total Revenue")
plt.show()
```
