```python
#LOADING THE DATASET:
import pandas as pd
df=pd.read_csv("sales.csv",sep='\t')
print(df.describe())
print(df.info())
print(df.head(5))
```

```python
#HANDLING MISSING VALUES:
print(df.isnull().sum)
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#TOTAL REVENUE:

print(df.columns)
total_sales=df.groupby("category")["total_price"].sum()
print(total_sales)
```

```python
#SORTING
sorted_df=df.sort_values(by=['category', 'total_price'], ascending=[True, False])
print(sorted_df)
```

```python
#CORRELATION :

correlation=df.corr(numeric_only=True)
print(correlation)
```
